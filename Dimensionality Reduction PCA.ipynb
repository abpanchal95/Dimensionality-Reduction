```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generating three features X1,X2 and X3 (random data)
X1 = np.random.normal(0, 1, size=10)
X2 = np.random.normal(0, 1, size=10)
X3 = np.random.normal(0, 1, size=10)
```

```python
#creating 3-dimensional X from X1,X2 and X3
X = np.array(list(zip(X1,X2,X3)))
print(X1[0], X2[0], X3[0], X[0])
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, X3)
plt.show()
```

```python
#importing PCA from sklearn
from sklearn.decomposition import PCA
```

```python
#creating PCA instance with n_components=2 (2-D)
pca_2d = PCA(n_components=2)
```

```python
#transforming X from 3-dimensions to 2-dimensions using PCA
Z_2d = pca_2d.fit_transform(X)
```

```python
#visualizing 2-dimensional Z_2d
%matplotlib
plt.scatter(Z_2d[:, 0], Z_2d[:, 1])
```

```python
#calculating retained variance from X to Z_2d
pca_2d.explained_variance_ratio_.cumsum()
```

### Here retained variance for n_components=2 is 92.26% which is good enough for data compression and train model on this new 2-dimensional data and for n_components=1 it is 53.36% which is not good enough it should be around 90% or above. So in conclusion for this data we can compress our data upto 2-dimension(lowest dimension) and for data visualization we don't need to depend on variance. For data visualization we can set n_components to be 1,2 or 3.
